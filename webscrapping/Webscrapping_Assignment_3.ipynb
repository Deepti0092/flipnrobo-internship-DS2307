# Ques2.In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.


```python
# creating empty list
product_URL=[]


for page in range(0,3):
    
#     brands=web_driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]')#scraping brands name by class name='_2WkVRV'
#     for i in brands:
#         brand.append(i.text)#appending the text in Brand list
  
           

    
    url1=web_driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']")
    for t in url1:
        product_URL.append(t.get_attribute('href'))
    next_button=web_driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    next_button.click()
    

        
    time.sleep(2)
  # for scraping expected delivery

           
  # for scraping availability
#     pr_avail=web_driver.find_elements_by_xpath('//span[@class="a-color-base a-text-bold"]')
#     for q in pr_avail:
#         avail.append(q.text)
 
```

```python
len(product_URL)
```

```python
print(product_URL)
```

```python
Brand=[]
name_pr=[]
price=[]
return_exch=[]
exp_del=[]
avail=[]

for url1 in product_URL:
    web_driver.get(url1)
    time.sleep(2)
    
    try:
        brands=web_driver.find_elements_by_xpath('//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        if len(brands):
             Brand.append(brands[0].text)
        else:
             Brand.append("-")
    
    except NoSuchElementException:
        Brand.append("-")
      
    try:
        name_product=web_driver.find_elements_by_xpath('//span[@id="productTitle" and @class="a-size-large product-title-word-break"]')#scraping name_product from the xpath
        print(name_product)
        name_pr.append(name_product[0].text)# appending the text in name_pr
    except NoSuchElementException:
        name_pr.append("-")
        
    try:   
        prices=web_driver.find_elements_by_xpath('//span[@class="a-price-whole"]') 
        price.append(prices[0].text)# appending the text in price list
    except NoSuchElementException:
        price.append("-")
     
    try:
        return_ex= web_driver.find_elements_by_xpath('//span[@class="a-size-small a-color-link a-text-normal"]')
        return_exch.append(return_ex[0].text)
    except NoSuchElementException:
        return_exch.append("-")
        
    try:
        expected_del=web_driver.find_elements_by_xpath('//span[@class="a-text-bold"]')
        exp_del.append(expected_del[0].text)
    except NoSuchElementException:
        exp_del.append("-")
        
        
    try:
        stock=web_driver.find_elements_by_xpath('//*[@id="availability"]/span')
        avail.append(stock[0].text)
    except NoSuchElementException:
        avail.append("-")
        

  
```

```python
len(Brand),len(price),len(name_pr),len(product_URL),len(exp_del),len(avail),len(return_exch)
```

```python
guitar_dic={'Brand_name':Brand,
            'Product_name':name_pr,
            'Price':price,
            'Return/Exchange':return_exch,
            'Expected_del':exp_del,
            'Availabity':avail,
            'Product_URL':product_URL}

df_guitar=dict([ (k,pd.Series(v)) for k,v in guitar_dic.items()])
df_guitar=pd.DataFrame(df_guitar)
df_guitar.head(60)
```
